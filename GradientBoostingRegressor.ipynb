```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor    #REGRESSION MEAN Numerical Number.
from sklearn.metrics import accuracy_score,mean_squared_error
from sklearn.preprocessing import OneHotEncoder
```

```python
#problem:
#create a predictive model using gradient boosting to forcast housing prices based on various features such as square footage,number of bedroom, no of bathrooms
# and locations (Real_life Model).
```

```python
data={'sqft':[1500,2000,1200,1800,1350],
     'Bed_room':[3,4,2,3,3],
     'wash_room':[2,2.5,1.5,2,2],
     'location':["Subrub","City","Rural","City","Subrub"],
     'prices':[250000,300000,180000,280000,220000]}
df=pd.DataFrame(data)
df
```

```python

Ohe=OneHotEncoder(sparse_output=False)
df[["Subrub","City","Rural"]]=Ohe.fit_transform(df[["location"]])
df.drop(columns=["location"],inplace=True)
```

```python
x=df.iloc[:,[0,1,2,4,5,6]]
y=df.iloc[:,[3]]
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
Model=GradientBoostingRegressor()
Model.fit(x_train,y_train)
```

```python
y_pred=Model.predict(x_test)
```

```python
#mean square error
mse=mean_squared_error(y_test,y_pred)
print("Mean_Squared_Eroor",mse)
```

```python
res=Model.predict([[1500,2,2,1,0,0]])
print(f"Predicted price is",res[0])
```
