```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

#Problem: A retailer businees owner ,you want to understand categories your customers into distinct segments based on their purchasing behvaviour.This
# segmentataion will help you tailor marketing starategies  and promotion of diffrent customer groups ultimately maximum sales and customer satisfaction.
#Dolouton:Using Gaussian matrix Model

#Ceate_Dataset
np.random.seed(42) #re-productibility
Segments_1=np.random.normal(loc=30,scale=5,size=100)   #Gaussin_distribution loc is mean and scale is standard diviation size is sample
Segment_2=np.random.normal(loc=60,scale=10,size=150)
Segment_3=np.random.normal(loc=90,scale=15,size=120)
data=np.concatenate([Segments_1,Segment_2,Segment_3]).reshape(-1,1)   #reshape into vetor
```

```python
scalar=StandardScaler()
scaled_data=scalar.fit_transform(data)
```

```python
n_components=3
GMM=GaussianMixture(n_components=n_components,random_state=42)
GMM.fit(scaled_data)
```

```python
#pridiction
cluster=GMM.predict(scaled_data)
```

```python
#visulaize
plt.scatter(data,np.zeros_like(data),c=cluster,cmap="viridis")
plt.title("Customer Segmentattion")
plt.xlabel("Puchased_Amount")
plt.show()
```

```python
#User_testing
user_input=float(input("Enter the purchase amount to predict the customer segments"))
scaled=scalar.fit_transform(np.array([[user_input]]))
result=GMM.predict(scaled).reshape(-1,1)[0]
print(f"The prediction  result is: {result}")
```
