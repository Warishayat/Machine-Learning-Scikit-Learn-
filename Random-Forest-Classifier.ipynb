```python
# Problems: you need to use random forest classifer to predict wheter a person is purchase to a produxt on the certain feature
# like age gender and estimated salery.
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
#Dummy data
data={"age":[25,35,45,20,30,50,40,55,60,35],
     "Gender":["male","female","male","male","female","female","male","female","male","female"],
     "Est-salery":[30000,40000,60000,20000,35000,80000,45000,90000,100000,50000],
     "purchase":[0,0,1,0,1,1,0,1,1,0]} #1 FOR PURCHASE 0 FOR NOT
```

```python
dataset=pd.DataFrame(data)
```

```python
dataset
```

```python
lE=LabelEncoder()
```

```python
Values=lE.fit_transform(dataset[["Gender"]])
dataset["Gender"]=Values  #1 for male and 0 for fe-male
dataset
```

```python
x=dataset[["age","Gender","Est-salery"]]
y=dataset["purchase"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rf_classifier=RandomForestClassifier(n_estimators=100,random_state=42) #n_estimator mean number of trees
```

```python
rf_classifier.fit(x_train,y_train)
```

```python
y_pred=rf_classifier.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy Model:",accuracy)
```

```python
age=int(input("Enter age :"))
gender=input("1 for male 0 for female")
salery=int(input("Enter salery"))
userdata=[[age,gender,salery]]
result=rf_classifier.predict(userdata)
if result[0]==1:
    print("Customer will likely to buy the product")
else:
    print("Customer will unlikely not buy the product")
```
