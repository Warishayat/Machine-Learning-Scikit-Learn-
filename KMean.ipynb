```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.cluster import KMeans

#A mall want to understand its customer base better to improve its marketing strateg.the goal is to cluster customer into group 
#based on the annual income and spending score,so the market team can tailor their campaign to each clustter preferance and behaviour.
#cluuster mean divide the data into the group

np.random.seed(42)
var={
    "Annual_income":np.random.randint(30000,100000,100),
    "Spending_Score":np.random.randint(1,100,100)
}
df=pd.DataFrame(var)
df.sample(7)

sns.scatterplot(data=df,x=df["Annual_income"],y=df["Spending_Score"]) #to visulize the data 

#we used clustter to group the people.

x=df.values
kmeans=KMeans(n_clusters=3,random_state=42) #random state for productivity
df["Cluster"]=kmeans.fit_predict(x)

#now visualize with cluster
sns.scatterplot(data=df,x=df["Annual_income"],y=df["Spending_Score"],hue=df["Cluster"])
#User_input
Annual_income=int(input("Enter the Annual_Income:"))
Spending_score=int(input("Enter Spending_Score."))
user_input={
    "Annual_income":[Annual_income],
    "Spending_score":[Spending_score]
}

user_df=pd.DataFrame(user_input)
result=kmeans.predict(user_df)
if result==1:
    print(f"The user belong to clustter {result} to visualise {sns.scatterplot(data=df,x=df["Annual_income"],y=df["Spending_Score"],hue=df["Cluster"])
}")
elif result==2:
    print(f"The user belong to clustter {result} to visualise {sns.scatterplot(data=df,x=df["Annual_income"],y=df["Spending_Score"],hue=df["Cluster"])
}")
elif result==0:
    print(f"The user belong to clustter {result} to visualise {sns.scatterplot(data=df,x=df["Annual_income"],y=df["Spending_Score"],hue=df["Cluster"])
}")

```
