```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC    #SUPPOR VECTOR CLASSIFIER
from sklearn.metrics import accuracy_score,classification_report
import seaborn as sns
```

```python
#Support vector machine (CLASSIFICATION-Model)
```

```python
#preperation of the data
data=pd.DataFrame({"age":[30,24,25,26,27,28,12,14],
                  "Monthly":[120,111,124,222,654,100,190,198],
                  "Churn":[1,0,0,1,1,1,0,1]})
```

```python
data
```

```python
x=data[["age","Monthly"]]
```

```python
y=data[["Churn"]]
```

```python
x,y
```

```python
sns.relplot(data=data,x="age",y="Churn",color="Black")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
Svm=SVC(kernel="linear",C=1.0)     #c is default regularization
```

```python
Svm.fit(x_train,y_train)
```

```python
y_pred=Svm.predict(x_test)   #predict the 20% data that is for prediction
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(f"Acuuracy Score is: {accuracy}")
```

```python
report=classification_report(y_test,y_pred)
print(f"Classification report is: {report}")
```

```python
age=int(input("Enter the age:"))
Monthly=int(input("Enter Monthly charge"))
score=Svm.predict([[age,Monthly]])
if score[0]==1:
    print("Customer will buy their churn")
else:
    print("Customer is not intrested")
```
