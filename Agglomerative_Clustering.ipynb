```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import AgglomerativeClustering
```

```python
#Problem:You are the data scientist in a retail company and the amrketing team is looking optimize their promotional campaigns by targeting specific 
#Customer segment the goal is to identify distinct groups of customer and based on their demographics information suc as age,income and spending score.
#Solution: AgglomerativeClustering Hirarichal
```

```python
np.random.seed(42)
data={"age":np.random.randint(18,65,100),
     "income":np.random.randint(30000,100000,100),
     "Spending_score":np.random.randint(1,100,100)}
data=pd.DataFrame(data)
```

```python
data
```

```python
sns.scatterplot(data=data,x="age",y="income")
```

```python
sc=StandardScaler()
```

```python
sc_data=sc.fit_transform(data)
clusters=AgglomerativeClustering(n_clusters=3,linkage="ward")  #to join 2 group we use wards.
```

```python
data["cluster"]=clusters.fit_predict(sc_data)
```

```python
sns.scatterplot(data=data,x="age",y="Spending_score",hue="cluster")
```
